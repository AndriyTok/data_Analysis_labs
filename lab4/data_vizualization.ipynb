```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
excel_file = pd.ExcelFile('medical_3.xlsx')
sheet_names = excel_file.sheet_names
print(f"{sheet_names}")
```

```python
data_glucose = pd.read_excel('medical_3.xlsx', sheet_name='Глюкоза')
data_glucose.head(20)
```

```python
data_sickness = pd.read_excel('medical_3.xlsx', sheet_name='Захворювання')
data_sickness.head()
```

```python
data_age_groups = pd.read_excel('medical_3.xlsx', sheet_name='Вікові групи')
data_age_groups.head()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Лінійна діаграма')
plt.plot(data_glucose['День'], data_glucose['Рівень цукру (мг/дл)'])
plt.grid()
plt.xlabel('День')
plt.ylabel('Рівень цукру (мг/дл)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Стовпчикова діаграма')
plt.bar(data_sickness['Захворювання'], data_sickness['Кількість пацієнтів'])
plt.grid()
plt.xlabel('Захворювання')
plt.ylabel('Рівень цукру (мг/дл)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title('Кругова діаграма')
plt.pie(data_age_groups['Кількість пацієнтів'], labels=data_age_groups['Вікова група'], autopct='%1.1f%%')
plt.show()
```

```python
# Функція
def f(x):
    return np.cos(x) * np.exp(np.abs(x))

# Інтервал та крок побудови
x = np.linspace(-4, 4, 400)
y = f(x)

# Побудова графіка
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='green')
plt.title('Графік функції cos(x) * e^|x|')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
